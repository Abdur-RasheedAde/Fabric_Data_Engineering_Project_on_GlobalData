```python
# Welcome to your new notebook
# Importing all relevant libraries
from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
from datetime import date
```

```python
#initialize the url and User-Agent in 2 variables; url and headers respectively 

url = 'https://yamahamusicstore.in/yamaha-standard-plus-pacs-12-shell-white-electric-guitar-gig-bag-included.html'

headers = ({"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36"})
```

```python
#request data from url using another variable page

page = requests.get(url, headers=headers)
page
```

```python
#check the data from page
#page.content
```

```python
#create a soup of data to organise data in html tags

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#Check datatype of content on page
type(page.content)
```

```python
#check the html code in soup
#print(soup)
```

```python
#getting the precise data from the soup
#inspect the page by right click on the page and click the top left icon the code pane to start inspection
#click on the exact item on the page (price) to see the html tag used for it

price = soup.find("span", class_="price").get_text()
print(price)
```

```python
#remove all non-digits from the price

price_num = "".join([spam for spam in price if spam.isdigit()])
print(price_num)
```

```python
#Get data for Reporting

today = date.today()
timetoday = datetime.datetime.now()
print("Today's date is: ", today, " ", type(today))
print("Time of the day is: ", timetoday, " ", type(timetoday))
```

```python
#Tabulate html data 
from pyspark.sql import Row
df = spark.createDataFrame([
    Row(Date = today, Datetime = timetoday, Price = price_num)
])
```

```python
#display data
display(df)
```

```python
#change Price column to integer

from pyspark.sql.types import IntegerType
df2 = df.withColumn("Price", df.Price.cast(IntegerType()))
df2.printSchema()
```
