```python
#define data path
dp1 = "Files/countriesandcontinents.csv"

#Load JSON data into a DataFrame
continent = spark.read.csv(dp1, header='true')
display(continent.limit(4))
```

```python
#Change Algerie to Algaria and USA to United States 
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when

# Initialize Spark session
spark = SparkSession.builder.appName("ReplaceRecord").getOrCreate()

# Replace 'Algerie' with 'Algeria' 
continent = continent.withColumn("Country", when(col("Country") == "Algerie", "Algeria").otherwise(col("Country")))

# Show updated DataFrame to verify the replacement
display(continent.limit(5))
```

```python
# Initialize Spark session
spark = SparkSession.builder.appName("ReplaceRecord2").getOrCreate()

# Replace 'USA' with 'United States' 
continent = continent.withColumn("Country", when(col("Country") == "USA", "United States").otherwise(col("Country")))
display(continent.limit(5))
```

```python
# Initialize Spark session
spark = SparkSession.builder.appName("ReplaceRecord3").getOrCreate()

# Replace 'Algerie' with 'Algeria' 
continent = continent.withColumn("Country", when(col("Country") == "U.A.E.", "United Arab Emirates").otherwise(col("Country")))
display(continent.limit(5))
```

```python
#Load JSON data into a DataFrame

df = spark.read.option("multiline","true").json("Files/continent-populations.json")
display(df)
```

```python
# round density and worldPop to 2 decimal places
from pyspark.sql.functions import round

df = df.withColumn("density", round(df["density"], 2))
df = df.withColumn("worldPop", round(df["worldPop"], 2))
display(df)
```

```python
#remove worldPop column

df = df.drop('worldPop')
display(df)
```

```python
#Change Oceania to Australia in df table
from pyspark.sql import functions as F

df = df.withColumn(
    "continent", F.when(F.col("continent") == "Oceania", "Australia")
    .otherwise(F.col("continent")))

display(df)
```

```python
#Change South/Central America to South America in continent table

continent = continent.withColumn(
    "Continent", F.when(F.col("Continent") == "South/Central America", "South America")
    .otherwise(F.col("Continent")))

display(continent.limit(5))
```

```python
#write dataframe to LakeHouse Table

df.write.format("delta").mode("overwrite").save("Tables/Continentpopulation")
```

```python
#write dataframe to LakeHouse Table

continent.write.format("delta").mode("overwrite").save("Tables/WorldContinent")
```
